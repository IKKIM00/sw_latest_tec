# 1. 정규 분포화 실습


```
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/gdrive')
```

```
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%matplotlib inline
```

```
data = pd.read_csv("/content/gdrive/My Drive/Yunjoong.ElementrySchool.2015.csv", names=['날짜', '모기수'])
train_data, test_data = train_test_split(data, test_size = 0.25, random_state=42)
mosq = pd.DataFrame(train_data['모기수'])
date = pd.DataFrame(train_data['날짜'])
mosq_scaled = StandardScaler().fit_transform(mosq)

row_n =["정규분포값"]
df = pd.DataFrame(mosq_scaled,index = date, columns=row_n)
```

```
df.sort_index()
```

```
df.plot.kde()
```

# 2. PCA 실습

```
pca = PCA().fit(mosq)
cpts = pd.DataFrame(pca.transform(mosq),index=date, columns=row_n)
cpts.sort_index()
```

```
cpts.plot.kde()
```
