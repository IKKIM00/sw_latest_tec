```
rf = pd.read_csv("/content/gdrive/My Drive/rainfall.2015.dat", sep='\t', index_col=[0])
rf = np.array(rf)
rf = rf.T
acc = 0
lst_acc = []
for i in range(len(rf)):
  for k in range(len(rf[i])):
    if(rf[i][k] == ' '):
      continue
    else:
      acc += float(rf[i][k])
  lst_acc.append(acc)
  acc=0
col = ["accumlation of rainfall"]
rf_df = pd.DataFrame(lst_acc, index=months, columns=col)
rf_df.plot()
```

### 상단의 두 그래프는 2015년 강수량을 측정한 그래프로 다음과 같은 특징들을 보여준다.

*   첫 번째 그래프는 매달 누적 강수량을 계산해서 어떤 달의 강수량이 가장 많은지를 알기 위해서 만들었다. 이를 통해서 가장 많은 강수량은 여름에 보여지는 것을 알 수 있고 상대적으로 겨울인 11~2월이 가장 강수량이 낮다는 것을 알 수 있다
*   두 번째 그래프는 데이터가 없는 날을 제외하고 강수량의 최대, 중위, 최소 값을 알기위해서 표현한 박스 그래프이다. 수염의 끝부분은 최대값과 최소값을 보여주는데 최소값은 -값이 존재할 수 없기 때문에 모두 0으로 보여진다. 하지만 몇몇 달에서 기록적인 강수량을 보여진 경우에는 이상치로 기록되어서 데이터상에 포함되지 못하였다.




```
avg_t = pd.read_csv("/content/gdrive/My Drive/temperature.avg.2015.dat", sep='\t', index_col=[0])
avg_t = np.array(avg_t)
avg_t = avg_t.T
acc = 0
lst_acc = []
for i in range(len(avg_t)):
  idx = 0
  for k in range(len(avg_t[i])):
    
    if(avg_t[i][k] == ' '):
      continue
    else:
      acc += float(avg_t[i][k])
      idx += 1
  acc /= idx
  lst_acc.append(acc)
  acc=0
col = ["temperature average"]
avgt_df = pd.DataFrame(lst_acc, index=months, columns=col)
avgt_df.plot()
```

```
avg_t = pd.read_csv("/content/gdrive/My Drive/temperature.avg.2015.dat", sep='\t', index_col=[0])
avg_t = np.array(avg_t)
avg_t = avg_t.T
s_lst = []
n_lst = []
for i in range(len(avg_t)):
  for k in range(len(avg_t[i])):
    if(avg_t[i][k] == ' '):
      continue
    else:
      s_lst.append(float(avg_t[i][k]))
  n_lst.append(s_lst)
  s_lst = []
d = pd.DataFrame(n_lst)
d1 = d.T
d1.columns = months
d1.boxplot()
```

### 위의 두 그래프는 2015년 연평균기온을 나타낸 것으로 다음과 같은 특징들을 보여준다.


*   첫번째 그래프는 매달 기록되어진 기온들을 모두 더하고 이를 기록된 날만큼 나누어서 연평균기온의 평균을 그래프로 나타낸 것이다
*   두 번째 그래프는 매달 평균기온의 최대, 최소, 중위값을 알아보기 위해서 그린 박스 그래프이다. 위 강수량 그래프와 마찬가지로 온도의 급격한 변화를 보여준 날짜들의 경우 이상치로 기록되어 데이터에 포함되지 못하고 이상값으로 기록된 한계가 있다.
